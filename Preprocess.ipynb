```python
girl_indices = []
boy_indices = []

for index, row in df.iterrows():
    # Check if the sentence contains 'woman' and 'man'
    contains_girl = 'girl' in row[1]
    contains_boy = 'boy' in row[1]
    # Append index to the respective list based on conditions

    if contains_boy and not contains_girl:
        boy_indices.append(index)
    if contains_girl and not contains_boy:
        girl_indices.append(index)

#print("Indices for 'woman':", girl_indices)
#print("Indices for 'man':", boy_indices)

urls_girl = df.iloc[girl_indices, 0].tolist()
urls_boy = df.iloc[boy_indices, 0].tolist()
```

```python
# Liste d'URLs de GIFs à télécharger



def download_gif(url,name, folder='downloaded_gifs'):
    try:
        response = requests.get(url)
        response.raise_for_status()

        # Extraire le nom du fichier du lien
        file_extension = url.split(".")[-1]
        file_name = os.path.join(folder, f'gif_{name}.{file_extension}')

        # Assurez-vous que le dossier de téléchargement existe
        os.makedirs(folder, exist_ok=True)

        # Enregistrez le fichier GIF
        with open(file_name, 'wb') as gif_file:
            gif_file.write(response.content)

        # print(f'GIF téléchargé avec succès : {file_name}')

    except requests.exceptions.HTTPError as errh:
        print ("Erreur HTTP:",errh)
    except requests.exceptions.ConnectionError as errc:
        print ("Erreur de connexion:",errc)
    except requests.exceptions.Timeout as errt:
        print ("Erreur de délai d'attente:",errt)
    except requests.exceptions.RequestException as err:
        print ("Erreur lors de la requête:",err)



# Télécharger chaque GIF dans la liste
#for idx,gif_url in enumerate(urls_girl):
    #download_gif(gif_url,idx,folder = 'girl')
for idx,gif_url in enumerate(urls_boy):
    name = len(urls_girl)+idx
    download_gif(gif_url,name,folder = 'boy')
```
